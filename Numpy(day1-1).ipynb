```python
# 즉 가로 365, 세로 333 짜리 이미지이며, 한 픽셀에 광도(밝기(명도)) 값이 0~255까지 들어가 있음
# 255가 제일 밝고(흰색) 0이 제일 어두움(검은색)
# 컬러이미지(1단계) 안에 회색조(grayscale)(2단계)가 들어있는 구조임
# 컬러이미지는 3장의 이미지가 필요한 구조임, 채널들이 있음 R G B 채널임. r값이 기록된 픽셀, g 값이 기록된 픽셀, b값이 기록된 픽셀 별로
# 각각 0~255 까지의 밝기 값이 들어있고
# 마지막 alpha 알파채널에 투명도를 나타내는 값이 들어있음

# 그래서 한 픽셀당 4개의 값이 필요하므로 3차원이 됨

# 위에서 0,0,0 위치의 픽셀값이 나온 이유는 그 픽셀의 각 채널 값 / 255 한 값이 나오는 거임

# 결국 4가 가지고 있는 의미는 한 픽셀이 4개의 정보를 가지고 있다는 뜻임

# 이게 가로로 365개가 있는거고, 이게 다시 세로로 333이 있는 거임
```
